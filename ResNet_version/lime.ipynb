```python
import numpy as np
from PIL import Image
from lime import lime_image
from skimage.segmentation import mark_boundaries
import matplotlib.pyplot as plt

# Load the image
image_path = r"D:\thesis\code\bangla_Handwritten_data\Test\172\bcc000000.bmp"
image = Image.open(image_path)

# Convert to numpy array
image_array = np.array(image)

# Create LIME explainer for image classification
explainer = lime_image.LimeImageExplainer()

# Define a function that returns predictions for your model (replace with your model)
def predict_fn(images):
    # Placeholder function - replace this with your actual prediction function
    # Example: return random predictions
    return np.random.rand(len(images), 1)  # Assuming binary classification

# Generate explanation
explanation = explainer.explain_instance(image_array, predict_fn, top_labels=1, hide_color=0, num_samples=1000)

# Show the explanation
temp, mask = explanation.get_image_and_mask(explanation.top_labels[0], positive_only=True, num_features=5, hide_rest=True)
img_boundry = mark_boundaries(temp / 2 + 0.5, mask)

# Display the original image and the LIME explanation
plt.figure(figsize=(12, 6))

# Plot original image
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')

# Plot LIME explanation
plt.subplot(1, 2, 2)
plt.imshow(img_boundry)
plt.title('LIME Explanation')

plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# List of image paths
image_paths = [
    
    r"D:\thesis\code\bangla_Handwritten_data\Test\172\bcc000000.bmp",
    r"D:\thesis\code\bangla_Handwritten_data\Test\173\bcc000058.bmp",
    r"D:\thesis\code\bangla_Handwritten_data\Test\183\bcc000036.bmp",
    r"D:\thesis\code\bangla_Handwritten_data\Test\188\bcc000024.bmp",
    r"D:\thesis\code\bangla_Handwritten_data\Test\192\bcc000018.bmp",
    r"D:\thesis\code\bangla_Handwritten_data\Train\220\bcc000012.bmp",
    r"D:\thesis\code\bangla_Handwritten_data\Train\221\bcc000009.bmp"
]

# Create a list to store loaded images
images = []
for path in image_paths:
    # Open each image using PIL
    img = Image.open(path)
    images.append(img)

# Calculate the total width and height for the combined figure
total_width = sum(img.width for img in images)
max_height = max(img.height for img in images)

# Create a new blank image to combine all images
combined_image = Image.new('RGB', (total_width, max_height))

# Paste each image into the combined image
x_offset = 0
for img in images:
    combined_image.paste(img, (x_offset, 0))
    x_offset += img.width

# Display the combined image using matplotlib
plt.imshow(combined_image)
plt.axis('off')  # Turn off axis
plt.show()
```
