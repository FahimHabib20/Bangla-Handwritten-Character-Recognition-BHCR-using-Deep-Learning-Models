```python
import numpy as np
from PIL import Image
from lime import lime_image
from skimage.segmentation import mark_boundaries
import matplotlib.pyplot as plt

# Load the image
image_path = r"D:\thesis\code\bangla_Handwritten_data\Test\172\bcc000000.bmp"
image = Image.open(image_path)

# Convert to numpy array
image_array = np.array(image)

# Create LIME explainer for image classification
explainer = lime_image.LimeImageExplainer()

# Define a function that returns predictions for your model (replace with your model)
def predict_fn(images):
    # Placeholder function - replace this with your actual prediction function
    # Example: return random predictions
    return np.random.rand(len(images), 1)  # Assuming binary classification

# Generate explanation
explanation = explainer.explain_instance(image_array, predict_fn, top_labels=1, hide_color=0, num_samples=1000)

# Show the explanation
temp, mask = explanation.get_image_and_mask(explanation.top_labels[0], positive_only=True, num_features=5, hide_rest=True)
img_boundry = mark_boundaries(temp / 2 + 0.5, mask)

# Display the original image and the LIME explanation
plt.figure(figsize=(12, 6))

# Plot original image
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')

# Plot LIME explanation
plt.subplot(1, 2, 2)
plt.imshow(img_boundry)
plt.title('LIME Explanation')

plt.show()
```

```python
import numpy as np
from PIL import Image
from lime import lime_image
from skimage.segmentation import mark_boundaries
import matplotlib.pyplot as plt

# Load the image
image_path = r"D:\thesis\code\bangla_Handwritten_data\Test\172\bcc000000.bmp"
image = Image.open(image_path)

# Convert to numpy array
image_array = np.array(image)

# Create LIME explainer for image classification
explainer = lime_image.LimeImageExplainer()

# Define a function that returns predictions for your model (replace with your model)
def predict_fn(images):
    # Placeholder function - replace this with your actual prediction function
    # Example: return random predictions for binary classification
    return np.random.rand(len(images), 1)  # Assuming binary classification

# Generate explanations for all predicted classes
explanations = []
num_labels = 10  # Specify the number of predicted labels (e.g., 10 for a 0-9 digit classifier)

for label in range(num_labels):
    # Define a function to predict the specified label
    def predict_label_fn(images, label=label):
        # Placeholder function - replace this with your model's prediction function
        # Example: return random predictions for the specified label
        return np.random.rand(len(images), 1)  # Assuming binary classification

    # Generate explanation for the specified label
    explanation = explainer.explain_instance(
        image_array, 
        predict_label_fn, 
        top_labels=1, 
        hide_color=0, 
        num_samples=1000
    )
    
    explanations.append(explanation)

# Display explanations for all predicted classes
plt.figure(figsize=(12, 6 * num_labels))

# Plot original image
plt.subplot(num_labels, 2, 1)
plt.imshow(image)
plt.title('Original Image')

# Plot LIME explanations for each label
for i, explanation in enumerate(explanations):
    temp, mask = explanation.get_image_and_mask(
        explanation.top_labels[0], 
        positive_only=True, 
        num_features=5, 
        hide_rest=True
    )
    img_boundry = mark_boundaries(temp / 2 + 0.5, mask)
    
    plt.subplot(num_labels, 2, 2 + 2*i)
    plt.imshow(img_boundry)
    plt.title(f'LIME Explanation - Label {i}')

plt.tight_layout()
plt.show()
```
